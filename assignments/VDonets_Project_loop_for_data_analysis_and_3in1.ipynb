```python
import os
import sys
import numpy
import matplotlib
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
#Reading Data
data_direc = (r"C:\\Users\\Atlas\\Desktop\\Data_for_Python\\CONTRAST\\")

nc_file = Dataset(data_direc+'CONTRAST-AWAS-MERGE_GV_R6_COMPOSITE.nc')
```

```python
print(nc_file.variables.keys())
```

```python
time = nc_file.variables['Time'][:]
CH3Cl = nc_file.variables['CH3Cl_Methyl_chloride'][:]

plt.scatter(time,CH3Cl,color='red')
plt.xlabel('Time')
plt.ylabel('Methyl Chloride')
plt.show()
```

```python
time = nc_file.variables['Time'][:]

for key in nc_file.variables:
    if key != 'Time':
        value = nc_file.variables[key][:]
        if value.shape == time.shape:
            fig = plt.figure()
            plt.rcParams.update({'font.size': 22})
            fig.set_size_inches(20,8)
            ax = plt.Axes(fig, [0., 0., 1., 1.], )
            ax=fig.add_subplot(111)
            plt.scatter(time,value)
            plt.xlabel('Time')
            plt.ylabel(key)
            fig.savefig('Time_vs_'+key+'.png')
            plt.close()
        else: print ('Not the same size as time')
```

```python
time = nc_file.variables['Time'][:]
butane = nc_file.variables['C4H10_nButane'][:]
pentane = nc_file.variables['C5H12_nPentane'][:]
benzene = nc_file.variables['C6H6_Benzene'][:]

fig = plt.figure()
plt.rcParams.update({'font.size': 22})
fig.set_size_inches(20,8)
ax = plt.Axes(fig, [0., 0., 1., 1.], )

ax=fig.add_subplot(311)
plt.scatter(time,butane)
plt.xlabel('Time')
plt.ylabel('Butane')

ax=fig.add_subplot(312)
plt.scatter(time,pentane)
plt.xlabel('Time')
plt.ylabel('Pentane')

ax=fig.add_subplot(313)
plt.scatter(time,benzene)
plt.xlabel('Time')
plt.ylabel('Benzene')

plt.show()

fig.savefig('Time_vs_But_Pen_Ben.png')
plt.close()
```
